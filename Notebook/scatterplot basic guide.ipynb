<h2>Imports</h2>

The first thing that you need to do when making your scatterplot is to import the necessary libraries. You will need to import both matplotlib.pyplot as well as numpy. It is common practice to import numpy as np and matplotlib.pyplot as plt, but you may change np and plt to whatever you desire. Whatever you choose to import them as will be the prefix you will need to use when calling functions from numpy or matplotlib.pyplot.

```python
import numpy as np
import matplotlib.pyplot as plt
```

<h2>Data</h2>

Next, you will need to create two arrays that will serve as the data points for the x and y axis. Depending on your needs, this can either be random, for which numpy has numerous functions to return a random array of different number distributions (i.e. gamma, hypergeometric, standard normal, chisquared, etc), or arrays you will need to fill with your own data. If you need your own data, np.genfromtxt may be a useful command to generate what you need. You may find the documentation of this function <a href="https://docs.scipy.org/doc/numpy/reference/generated/numpy.genfromtxt.html">here.</a> If you need random numbers, a list of the different distributions you can make random samples from and the documentation can be found <a href="https://docs.scipy.org/doc/numpy-1.12.0/reference/routines.random.html">here.</a> When calling a random function, you must use np.random.function(parameters…) to do so instead of simply np.function(parameters…).


```python
x = np.random.randn(512)
y = np.random.randn(512)
```

<h2>Scatterplot</h2>

Next you will want to create the scatterplot using plt.scatter(x, y, s=number, c=see_below, marker=”marker”, cmap=”colormap”, alpha=number).
<ul>
    <li><b>x, y</b> – the arrays of input. They should be the same size.</li>
    <li><b>s</b> – size of the scatterplot points.</li>
    <li><b>c</b> – from the documentation: “c can be a single color format string, or a sequence of color specifications of length N, or a sequence of N numbers to be mapped to colors using the cmap and norm specified via kwargs (see below). Note that c should not be a single numeric RGB or RGBA sequence because that is indistinguishable from an array of values to be colormapped. c can be a 2-D array in which the rows are RGB or RGBA, however, including the case of a single row to specify the same color for all points.” To clarify: it's a list/array of color specifications equal in length to the length of the x and y data sets. For example, c=x+y will give each point a color in their color map based on the sum of their x and y coordinates. </li>
    <li><b>marker</b> – a string that determines the style of marker used for the points on the scatterplot. Click <a href="https://matplotlib.org/devdocs/api/markers_api.html">here.</a> for a full list of available marker styles.</li>
    <li><b>cmap</b> – a string that determines the color map to use. It’s only used if c is an array of floats. For a list of available colormaps, click <a href="https://matplotlib.org/examples/color/colormaps_reference.html">here.</a></li>
    <li><b>alpha</b> – a number between 0 and 1 that determines the transparency of a point on the plot. 0 would be invisible while 1 would be completely opaque.</li>
</ul>


```python
#c=x+y will give the points a color based of the sum of their x and y coords in the colormap hsv
plt.scatter(x, y, s=25, c=x+y, cmap="hsv", alpha=.25)
```

<h2>Plot Details</h2>

Next, you may add miscellaneous things to your graph, such as a title, axes labels, annotations, and so on. While a full list of all the things you can do are documented <a href="https://matplotlib.org/api/pyplot_api.html">here,</a>
Some of the more commonly needed commands are…

<ul>
    <li><b>plt.title(name)</b> puts a title equal to the name parameter at the top of the graph.</li>
    <li><b>plt.x(y)label(label)</b> puts a label on the x(y)-axis equal to the label parameter.</li>
    <li><b>plt.x(y)ticks([ticks],[labels])</b> will set ticks at the values given in the array ticks with the labels at the equivalent index in the array labels. Special characters (for example, the pi symbol) may be used. To use a special character, it must be formatted as `r’$\symbol$’` . For example, to write -π/2 as the label for the first tick, you would put `r’$-\pi/2$’` as the first element in the labels array. By default for scatterplots, the ticks are based off the size of the scatterplot’s data values, but is generally something reasonable.</li>
    <li><b>plt.x(y)lim(min, max)</b> will set the minimum and maximum limits for that axis on the graph to the values min and max. Scatterplots by default set the limit to show every point in the scatterplot and nothing else.</li>
    <li><b>plt.annotate(label, xy=(x, y), xytext=(x,y), arrowprops=dict(property1=value, ...)</b> will put an annotation with a label given by the label parameter. The label will be placed at the coordinates given by xytext with an arrow pointing to the coordinates given by xy. arrowprops is a dictionary of properties that will be given to the arrow the annotation uses. If arrowprops is None (which is the default), or its dictionary is empty, then no arrow will appear. A full list of arrowprops's parameters may be found in the link at the bottom of this section. When using this function, you must give the parameters by writing out parameter=(information). or else the function will not understand what parameter you’re trying to pass info to. Information on other parameters for this function can be found <a href="https://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.annotate">here.</a></li>
</ul>


```python
plt.title("Test Scatterplot")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.xticks(np.arange(-5,6))
plt.yticks(np.arange(-5,6))
plt.xlim(-5,5)
plt.ylim(-5,5)
plt.annotate("This is the last point graphed", xy=(x[-1], y[-1]), xytext=(x[-1]+1,y[-1]+1), arrowprops=dict(width=1,color="red",alpha=.5))
```

<h2>Displaying your scatterplot</h2>

Finally, call plt.show() to display your scatterplot.

```python
plt.show()
```
